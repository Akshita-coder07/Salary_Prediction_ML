Feature Engineering
Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
Label_Encoder = LabelEncoder()
```

```python
df1['Gender_Encode'] = Label_Encoder.fit_transform(df1['Gender'])
df1.head(10)
```

```python
df1['Degree_Encode'] = Label_Encoder.fit_transform(df1['Degree'])
df1.head(10)
```

Data after label Encoding

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
df1['Age_scaled'] = std_scaler.fit_transform(df1[['Age']])
df1['Experience_years_scaled'] = std_scaler.fit_transform(df1[['Experience_years']])
df1.head()
```

Dependent and Independent 

```python
x = df1[['Age_scaled','Gender_Encode','Degree_Encode','Job Title_Encode','Experience_years_scaled']] #independennt
y = df1['Salary'] #dependent
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
y_train
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
x_train
```

Model Development

```python
from sklearn.linear_model import LinearRegression
Linear_regression_model = LinearRegression() 
```

```python
Linear_regression_model.fit(x_train, y_train) # model traning
```

```python
y_pred_lr = Linear_regression_model.predict(x_test) # model prediction
y_pred_lr
```

```python
# Convert y_test (Series) into DataFrame with column name 'Salary'
df1 = y_test.to_frame(name='Salary')

# Add predictions from y_pred_lr
df1['predicted'] = y_pred_lr

df1.head(10)
```

```python
plt.scatter(df1['Salary'],df1['predicted'],color='Green',marker="o",s=100)
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.show()
```

```python
print(len(x_train))
print(len(y_train))
```

```python
sns.distplot(y_pred_lr)
```

```python
# Assuming df1 has 'Salary' (actual) and 'predicted' (predictions)
df1['error'] = df1['Salary'] - df1['predicted']
df1['abs_error'] = abs(df1['error'])

# Mean Absolute Error (MAE)
mae = df1['abs_error'].mean()

# Mean Squared Error (MSE) & Root Mean Squared Error (RMSE)
df1['squared_error'] = df1['error'] ** 2
mse = df1['squared_error'].mean()
rmse = np.sqrt(mse)

# Mean Absolute Percentage Error (MAPE) & Mean Percentage Error (MPE)
percentage_error = (df1['error'] / df1['Salary']) * 100
absolute_percentage_error = abs(percentage_error)
mape = absolute_percentage_error.mean()
mpe = percentage_error.mean()

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("MAPE:", mape)
print("MPE:", mpe)
```
